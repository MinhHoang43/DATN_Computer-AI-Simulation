```python
from ultralytics import YOLO

model=YOLO("yolo11s.pt")
#Cell 5 - Load YOLO
```

```python
results=model.train(

    data="/content/PCB_Detection_YOLO/data.yaml",

    epochs=100,

    imgsz=640,

    batch=8,

    patience=50,

    workers=2,

    cache=True,

    project="/content/runs/detect",

    name="PCB_Detection_GAN_Result"

)
#Cell 6 - Train
```
